# Import Libraries

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plot
from pathlib import Path
```

# Load Images

```python
def get_images_by_dir(directory):
    Images = []
    Labels = []
    Paths = []
    labels = os.listdir(directory)
    for label in labels:
        category = labels.index(label)
        for image in os.listdir(Path(directory,label)):
            image_path = Path(directory, label, image)
            img_data = cv2.imread(f'{image_path}')
            Images.append(img_data)
            Labels.append(label)
            Paths.append(image_path)
    return Images, Labels, Paths

Images, Labels, Paths = get_images_by_dir("../data/seg_train")
```

# Display Images

```python
f,ax = plot.subplots(5,5) 
f.subplots_adjust(0,0,3,3)
for i in range(0,5,1):
    for j in range(0,5,1):
        rnd_number = np.random.randint(0,len(Images))
        ax[i,j].imshow(Images[rnd_number])
        ax[i,j].set_title(Labels[rnd_number])
        ax[i,j].axis('off')
```
