# Import Libraries

```python
import cv2
import numpy as np
import os
from PIL import Image
from pathlib import Path
```

# Load Images

```python
def get_images_by_dir(directory):
    Images = []
    Labels = []
    Paths = []
    labels = os.listdir(directory)
    for label in labels:
        category = labels.index(label)
        for image in os.listdir(Path(directory,label)):
            image_path = Path(directory, label, image)
            img_data = cv2.imread(f'{image_path}')
            Images.append(img_data)
            Labels.append(label)
            Paths.append(image_path)
    return Images, Labels, Paths

Images, Labels, Paths = get_images_by_dir("../data/seg_train")
```
